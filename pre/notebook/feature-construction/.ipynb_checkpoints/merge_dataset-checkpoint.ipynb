```python
import pandas as pd
from pandas import DataFrame
from pandas import Series

import numpy as np
import matplotlib.pyplot as plt
```

### 设置路径

```python
path_pre = '../../'
path_original_dataset = path_pre + 'original-dataset/'
path_intermediate_dataset = path_pre + 'intermediate-dataset/'
```

### 加载数据

```python
user = pd.read_hdf(path_intermediate_dataset + 'fg_user.h5')
user.head()
```

```python
ad = pd.read_hdf(path_intermediate_dataset + 'fg_ad.h5')
ad.head()
```

```python
action_context = pd.read_hdf(path_intermediate_dataset + 'fg_action_context.h5')
action_context.head()
```

### 合并为 dataset

```python
dataset = pd.merge(action_context, user, on='userID')
dataset = dataset.merge(ad, on='creativeID')
dataset.head()
```

```python
dataset['userID'].value_counts().head()
```

```python
dataset.loc[dataset['userID'] == 2453862, 'appPlatform'].value_counts()
```

通过简单的分析发现，不存在不匹配的情况

### 构造 is_pref_cat 特征

```python
dataset['is_pref_cat'] = dataset['appCategory'] == dataset['cat_pref']
dataset.head()
```

```python
dataset['is_pref_cat'].value_counts()
```

### 删除 creativeID, userID 两列

```python
del dataset['creativeID']
del dataset['userID']
dataset.head()
```
