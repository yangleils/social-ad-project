```python
import pandas as pd
from pandas import DataFrame
from pandas import Series

import numpy as np
import matplotlib.pyplot as plt
```

### 设置路径

```python
path_pre = '../'
path_original_dataset = path_pre + 'original-dataset/'
path_intermediate_dataset = path_pre + 'intermediate-dataset/'
```

### 加载数据

```python
user = pd.read_hdf(path_intermediate_dataset + 'user_fe.h5')
user.head()
```

```python
ad = pd.read_hdf(path_intermediate_dataset + 'ad_fe.h5')
ad.head()
```

```python
action_context = pd.read_hdf(path_intermediate_dataset + 'action_context_test_ol_fe.h5')
action_context.head()
```

### 合并为 testset_ol

```python
testset_ol = pd.merge(action_context, user, on='userID')
testset_ol = testset_ol.merge(ad, on='creativeID')
testset_ol.head()
```

### 删除 creativeID, userID 两列
